```python
from GoogleNews import GoogleNews

# 初始化 GoogleNews 物件
googlenews = GoogleNews()

# 搜尋關鍵字（例如：Trump）
googlenews.search('711')

# 取得搜尋結果
result = googlenews.result()

# 列印新聞數量
print(f"共有 {len(result)} 篇新聞")

# 列印每篇新聞的標題、日期、描述和連結
for n in range(len(result)):
    print(f"第 {n+1} 篇新聞")
    print("標題:", result[n]['title'])
    print("日期:", result[n]['date'])
    print("描述:", result[n]['desc'])
    print("連結:", result[n]['link'])
    print("\n")

df = pd.DataFrame()

# 將每篇新聞的標題、日期、描述和連結添加到 DataFrame
for n in range(len(result)):
    df = pd.concat([df, pd.DataFrame({'標題': [result[n]['title']], '日期': [result[n]['date']], '描述': [result[n]['desc']], '連結': [result[n]['link']]})], ignore_index=True)
```
